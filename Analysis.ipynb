<h4> What is the Customer Attrition Rate? </h4>

```python
num_of_attrited_customer = len(data[data['Attrition_Flag'] == 'Attrited Customer' ])
num_of_total_customer = data.shape[0]

attrition_rate = num_of_attrited_customer / num_of_total_customer
print(f'The number of attrited customer: {num_of_attrited_customer}')
print(f'The number of total customer: {num_of_total_customer}')
print(f'The attrition rate: {attrition_rate * 100}')
```

```python
y = pd.get_dummies(data['Attrition_Flag'])
y = y.drop(["Attrited Customer"], axis = 1)
y = y.rename(columns = {"Existing Customer": "Current Customer"})
y
```

```python
X = data.drop(columns = ['Attrition_Flag'])
X
```

<h4> Train Test Split </h4>


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 666)
```

<h4> Model Building </h4>

```python
model = RandomForestClassifier(random_state = 666)
#       LogisticRegression(solver='liblinear', random_state=0)
#       GaussianNB()
#       svm.SVC(kernel='linear') 
```

```python
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
joblib.dump(model, 'model.pkl')
```
